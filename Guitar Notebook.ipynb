# Acoustic vs Electric guitar classification

### Introduciton

#### In this notebook I would like to take you on a journey of image classificaition inspired by Jeremy howard - lead reasearcher of fast ai. In this notebook we will train a convolutional neural network (CNN) to learn and be able to tell the difference between an acoustic guitar and an electric guitar. While there are many hybrids, as a guitarist I've decided to declare hollowbody guitars with a soundhole as acoustic and a guitar with exposed pickups (designed for major amplificaiton) to be electrics. To do this I scraped google images for roughly 400 guitar images and then when through the process of data munging the images to ensure we had a quality set of acoustic and electric guitar images for our CNN to learn from. 

#### The findings of this research lend evidence to the fact that we can build a great classification system with a limited number of images (Less than a 200 of each category). This is quite impressive as we are able to make accurate inference on two objects which are very similar in that they serve near identical purposes.

```python
#here we take in our fast ai libraries
from fastai.vision import *
from fastai import *
```

# Setting up folders and pathing for data download

```python
#let's set up for our electric guitars
folder = 'electric_guitars'
file = 'urls_electric_guitars.csv'
```

```python
#we must set a path and allow the creation of a folder
path = Path('data/guitars')
dest = path/folder
dest.mkdir(parents=True, exist_ok=True)
```

```python
#Here is the command we will use to actually download the images into our directories
download_images(path/file, dest, max_pics=200)
```

```python
#Here we can see we've set up our directory
path.ls()
```

```python
#let's set up another directory for the acoustic guitars 
folder = 'acoustic_guitars'
file = 'urls_acoustic_guitars.csv'
```

```python
#again let's set a path and allow the creation of a folder to store our target acoustic guitars
path = Path('data/guitars')
dest = path/folder
#dest.mkdir(parents=True, exist_ok=True)
```

```python
#here we have our target classes
classes = ['electric_guitars','acoustic_guitars']
```

```python
classes
```

```python
#Again we will download the images into a directory
download_images(path/file, dest, max_pics=200)
```

```python
#Seems  we are setup now!
path.ls()
```

```python
#Let's also clean out all of the images that are corrupted or won't load
for c in classes:
    print(c)
    verify_images(path/c, delete=True, max_size=500)
```

# Let's View the Data

```python
#setting random seed so we can get consistent results
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
        ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
#Here we can see our classes  are set appropriately
data.classes
```

```python
#now let's actually take a look at our images 
data.show_batch(rows=3, figsize=(7,8))
```

```python
#Here we see we will train on 228 images and validate with 56 testing images
data.classes, data.c, len(data.train_ds), len(data.valid_ds)
```

# Awesome! now let's train :)

```python
#Let's start out with a resnet34 architecture
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
#Train our data here on 4 epochs
learn.fit_one_cycle(4)
```

```python
#save the model
learn.save('stage-1')
```

```python
#let's now see if we can beat that benchmark
learn.unfreeze()
```

```python
#here we will find our optimal learning rate for our data
learn.lr_find()
```

```python
#now we can see our learning rate and where it increases
learn.recorder.plot()
```

# Where are we getting most of our errors?

```python
#let's set up to take a look at the images which are giving us our worst performance
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
#here we can see our top offenders - seems our CNN doesn't like groups of guitars
interp.plot_top_losses(9, figsize=(15,11))
```

```python
#let's adjust for the the learning rate and see if we can improve
learn.fit_one_cycle(2, max_lr=slice(1e-5,1e-3))
```

```python
#error rate went up ??
learn.save('stage-2')
```

# Interpretation

```python
#let's now take a look at our confusion matrix from stage 2
learn.load('stage-2');
```

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

# Let's clean up our data

```python
from fastai.widgets import *
```

```python
db = (ImageList.from_folder(path)
                   .split_none()
                   .label_from_folder()
                   .transform(get_transforms(), size=224)
                   .databunch()
     )
```

```python
learn_cln = cnn_learner(db, models.resnet34, metrics=error_rate)

learn_cln.load('stage-2');
```

```python
ds, idxs = DatasetFormatter().from_toplosses(learn_cln)
```

```python
ImageCleaner(ds, idxs, path)
```

```python
#any duplicates?
ds, idxs = DatasetFormatter().from_similars(learn_cln)
```

```python
#let's clean em up!
ImageCleaner(ds, idxs, path, duplicates=True)
```

# Let's train again with our cleaned up data

```python
np.random.seed(42)
data = ImageDataBunch.from_csv(path, folder=".", valid_pct=0.2, csv_labels='cleaned.csv',
         ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
#reset our learn  variable
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
```

```python
#now let's train it
learn.fit_one_cycle(4)
```

```python
#Seem's likea big improvement!
learn.save('stage-1')
```

```python
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
#let's take a fresh look at our top offenders
interp.plot_top_losses(9, figsize=(15,11))
```

```python
#Our confusion matrix seems pretty good this time!
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

# Let's see if we can beat that with Resnet 50

```python
np.random.seed(42)
data = ImageDataBunch.from_csv(path, folder=".", valid_pct=0.2, csv_labels='cleaned.csv',
         ds_tfms=get_transforms(), size=224, num_workers=4).normalize(imagenet_stats)
```

```python
learn = cnn_learner(data, models.resnet50, metrics=error_rate)
```

```python
#doesn't seem as though we've improved here
learn.fit_one_cycle(4)
```

```python
#let's save our bench mark and see if we can improve with some tuning
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
learn.lr_find()
```

```python
#Let's again look at our learning rate
learn.recorder.plot()
```

```python
#let's again adjust for the the learning rate and see if we can improve
learn.fit_one_cycle(2, max_lr=slice(1e-5,1e-3))
```

```python
#Wow - we're up to 98 - 99% Accuracy!
learn.save('stage-2-Resnet50')
```

```python
#So we will take a look again at the images we are losing on
interp = ClassificationInterpretation.from_learner(learn)

losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```

```python
#let's take a fresh look at our top offenders
interp.plot_top_losses(9, figsize=(15,11))
```

```python
#now lets see our confusion matrix
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
